```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

```python
x = np.arange(11)
y1 = 2*x+3
y2 = 4*x+5
y3 = x**2
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,'r--o',label='Potato')
plt.plot(x,y2,'g:',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,label='Potato',color='r',linestyle='--',marker='*',
        linewidth = 2,markerfacecolor='b',ms=10)
plt.plot(x,y2,'g:',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.scatter(x,y1,label='Potato',color='r',marker='^')
plt.scatter(x,y2,label='onion',marker='*')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.array([1,2,3])*2
y1 = np.array([10,20,30])
y2 = np.array([13,22,25])

plt.bar(x,y1,width = 0.5,tick_label = ['Silver','Gold','Platinum'],label="Previous Year")
plt.bar(x+0.5,y2,width = 0.5,label="Current Year",linewidth=10)
plt.ylabel('Price')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
plt.pie([20,30,10,50,20],explode=(0,0,0,0.11,0),
        labels=['a','b','c','d','e'],shadow=True,labeldistance = 0.5)
plt.title('Pie Chart')
plt.show()
```

```python
# nda1 = np.random.normal(50,20,1000)
# nda2 = np.random.normal(70,10,1000)

sdn = np.random.randn(100)
nda1 = sdn*20+50
nda2 = sdn*10+70
```

```python
plt.hist(nda1,alpha = 0.5,label = "Class A")
plt.hist(nda2,alpha = 0.5,label = "Class B")

plt.xlim(0,100)
# plt.ylim(0,25)

plt.ylabel('Student Frequency')
plt.title("Student Marks")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3])*2
y1 = np.array([10,20,30])
y2 = np.array([13,22,25])

plt.barh(x,y1)
plt.barh(x+0.5,y2)
plt.title('Metal Prices')
# plt.legend()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
lines = plt.plot(x,y1,x,y2)
plt.setp(lines[0],label='Potato',color ='r',linestyle='--',marker='*',
        linewidth = 2,markerfacecolor='b',ms=10)
plt.setp(lines[1],'color','g',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(0,5,100)
y1 = np.sin(x*2*np.pi)
y2 = np.cos(2*np.pi*x)

plt.figure(figsize=(8,2))
plt.subplot(211)
plt.plot(x,y1)
plt.title('Sin Graph')
plt.grid()

plt.subplot(212)
plt.plot(x,y2)
plt.title('Cos Graph')
plt.grid()

plt.suptitle("Trignometeric Functions")
plt.subplots_adjust(top= 0.7,hspace=0.5)
plt.show()
```

```python
x = np.linspace(0,5,100)
y1 = np.sin(x*2*np.pi)
y2 = np.cos(2*np.pi*x)

plt.figure(figsize=(10,4))
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.plot(x,y1)
ax1.set_title('Sin Graph')
ax1.grid()

ax2.plot(x,y2)
ax2.set_title('Cos Graph')
ax2.grid()

plt.suptitle("Trignometeric Functions")
plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
x = np.linspace(0,5,100)
y1 = np.sin(x*2*np.pi)
y2 = np.cos(2*np.pi*x)

fig,ax = plt.subplots(2,2,figsize=(10,4))

ax[0,0].plot(x,y1)
ax[0,0].set_title('Sin Graph')
ax[0,0].grid()

ax[0,1].plot(x,y2)
ax[0,1].set_title('Cos Graph')
ax[0,1].grid()

fig.suptitle("Trignometeric Functions")
fig.subplots_adjust(hspace=0.5)
plt.show()
```
