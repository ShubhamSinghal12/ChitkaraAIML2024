```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

```python
x = np.arange(11)
y1 = 2*x+3
y2 = 4*x+5
y3 = x**2
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,'r--o',label='Potato')
plt.plot(x,y2,'g:',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,label='Potato',color='r',linestyle='--',marker='*',
        linewidth = 2,markerfacecolor='b',ms=10)
plt.plot(x,y2,'g:',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.scatter(x,y1,label='Potato',color='r',marker='^')
plt.scatter(x,y2,label='onion',marker='*')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.array([1,2,3])*2
y1 = np.array([10,20,30])
y2 = np.array([13,22,25])

plt.bar(x,y1,width = 0.5,tick_label = ['Silver','Gold','Platinum'],label="Previous Year")
plt.bar(x+0.5,y2,width = 0.5,label="Current Year",linewidth=10)
plt.ylabel('Price')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
plt.pie([20,30,10,50,20],explode=(0,0,0,0.11,0),
        labels=['a','b','c','d','e'],shadow=True,labeldistance = 0.5)
plt.title('Pie Chart')
plt.show()
```

```python
x1 = np.random.randint(30,70,100)
x2 = np.random.randint(50,90,100)
```

```python
plt.hist(x1,alpha = 0.5,label = "Class A")
plt.hist(x2,alpha = 0.5,label = "Class B")

plt.xlim(0,100)
plt.ylim(0,25)

plt.ylabel('Student Frequency')
plt.title("Student Marks")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3])*2
y1 = np.array([10,20,30])
y2 = np.array([13,22,25])

plt.barh(x,y1)
plt.barh(x+0.5,y2)
plt.title('Metal Prices')
# plt.legend()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
lines = plt.plot(x,y1,x,y2)
plt.setp(lines[0],label='Potato',color ='r',linestyle='--',marker='*',
        linewidth = 2,markerfacecolor='b',ms=10)
plt.setp(lines[1],'color','g',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
# plt.plot(x,y3)
plt.legend()
plt.grid()
plt.show()
```
