```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
X = np.array(random.sample(range(99999),1000))
```

```python
u = X.mean()
std = X.std()
X = (X-u)/std
Y = 2*X+11

X_Train,X_Test = X[:800],X[800:]
Y_Train,Y_Test = Y[:800],Y[800:]
```

```python
plt.scatter(X_Train,Y_Train)
```

```python
def hypothesis(X,theta):
    y_ = theta[0] + theta[1]*X
    return y_
# hypothesis(X,[11,2])
```

```python
def cost(X,theta,Y):
    N = X.shape[0]
    y_ = hypothesis(X,theta)
    er = np.sum((y_ - Y)**2)/N
    return er
# cost(X,[11,2],Y)
```

```python
def gradient(X,theta,Y):
    N = X.shape[0]
    y_ = hypothesis(X,theta)
    grad = np.zeros((2,))
    grad[0] = np.sum((y_ - Y))
    grad[1] = np.sum((y_ - Y)*X)
    return grad/N
# gradient(X,[1,1],Y)
```

```python
def gradient_descent(X,Y,lr = 0.001,epochs = 1000):
    theta = np.zeros((2,))
    costs = []
    for e in range(epochs):
        costs.append(cost(X,theta,Y))
        grad = gradient(X,theta,Y)
        for i in range(theta.shape[0]):
            theta[i] = theta[i] - lr*grad[i]
    return theta,costs
```

```python
theta,costs = gradient_descent(X_Train,Y_Train,0.003,5000)
theta
```

```python
plt.plot(costs)
```

```python
y_ = hypothesis(X_Test,theta)
for i in range(len(y_)):
    print(y_[i],Y_Test[i])
```
