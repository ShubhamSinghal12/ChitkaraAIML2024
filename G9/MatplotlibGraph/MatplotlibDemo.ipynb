```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

```python
x = np.arange(11)
y1 = 2*x+3
y2 = 4*x+5
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,'r--o',label='Potato')
plt.plot(x,y2,'g:^',label='onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title("Vegetable Prices",fontdict=font)

plt.legend()
plt.grid()
plt.show()
```
