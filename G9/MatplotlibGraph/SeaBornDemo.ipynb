```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.groupby('day').mean(True)
```

```python
sns.barplot(data=tips,y='day',x='total_bill',estimator='sum',errorbar=None,hue='sex',orient='h')
```

```python
sns.histplot(data = tips,x='total_bill',bins=30,kde=True)
```

```python
sns.kdeplot(data=tips,x='tip')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```
