```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.groupby('day').mean(True)
```

```python
sns.barplot(data=tips,y='day',x='total_bill',estimator='sum'
           ,hue='sex',orient='h')
```

```python
sns.histplot(data=tips,x='total_bill',bins=10,kde=True)
```

```python
sns.kdeplot(data = tips,x='total_bill',fill=True)
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='sex',markers=['*','^'],s=100)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(data=fmri,x='timepoint',y='signal',hue='event',style='region',markers=True)
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
sns.set_style('whitegrid')
sns.set_context('paper',font_scale=1.5)
sns.pairplot(tips,hue='day',kind='scatter',palette='seismic')
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,jitter=True)
```

```python
sns.violinplot(data=tips,x='day',y='total_bill')
sns.swarmplot(data=tips,x='day',y='total_bill',color='white')
```

```python
flights = sns.load_dataset('flights')
```

```python
fmat = flights.pivot(index='month',columns='year',values='passengers')
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(fmat,annot=True,cmap='Blues',fmt='g')
```
