```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')
```

```python
x = np.arange(11)
y1 = 2*x+3
y2 = 4*x+5
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,label='Potato',c = 'r',marker='^',linestyle='--',linewidth=5,markerfacecolor='blue', markersize=12)
plt.plot(x,y2,label='Onion')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
lines = plt.plot(x,y1,x,y2)
plt.setp(lines[0],label='Potato',c = 'r',marker='^',linestyle='--',linewidth=5,markerfacecolor='blue', markersize=12)
plt.setp(lines[1],label='Onion')

plt.legend()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
plt.scatter(x,y1,label='Potato',marker='^')
plt.scatter(x,y2,label='Onion',marker='*')
plt.xlabel('Quantity',fontdict=font)
plt.ylabel('Price',fontdict=font)
plt.title('Vegetable Prices',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.style.use('dark_background')
x = np.array([1,2,3])*1.5
y1 = np.array([10,20,30])
y2 = np.array([13,22,35])

plt.bar(x,y1,width = 0.5,label = 'Previous Year',
        tick_label=['Silver','Gold','Platinum'])
plt.bar(x+0.5,y2,width = 0.5,label = 'Current Year')
plt.ylabel('Price')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
plt.pie([10,20,10,40,10],labels=['a','b','c','d','e']
        ,explode = (0,0,0,0.1,0),shadow = True,autopct='%1.1f%%')
plt.show()
```

```python
val1 = np.random.randint(30,70,100)
val2 = np.random.randint(50,90,100)
```

```python
plt.hist(val1,histtype = 'barstacked',orientation= 'horizontal',alpha=0.5,linewidth=0.2, edgecolor="black",label='Class A')
plt.hist(val2,histtype = 'barstacked',orientation= 'horizontal',alpha=0.5,linewidth=0.2, edgecolor="black",label = 'Class B')
# plt.xlim(0,100)
# plt.ylim(0,25)

plt.xlabel('Buckets')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()
```
