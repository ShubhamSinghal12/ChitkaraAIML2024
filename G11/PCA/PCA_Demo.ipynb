```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = np.random.randint(1,100,5000).reshape((1000,5))
k = 2
```

```python
data = load_breast_cancer()
y = data.target
data = data.data
scale = StandardScaler()
data = scale.fit_transform(data)
k = 3
```

```python
data.shape
```

```python
covM = np.cov(data,rowvar=False)
```

```python
covM.shape
```

```python
eigenValues, eigenVector = np.linalg.eigh(covM)
```

```python
eigenValues, eigenValues.shape
```

```python
# eigenVector,
eigenVector.shape
```

```python
sort_index = np.argsort(eigenValues)[::-1]
```

```python
sort_index
```

```python
kBest = eigenVector[sort_index][:k,:]
```

```python
kBest.shape
```

```python
newData = (kBest.dot(data.T)).T
```

```python
newData.shape
```

```python
# newData
```

```python
plt.scatter(newData[:,0],newData[:,1],c=y)
plt.show()
```

```python
pca = PCA(n_components=3)
newData = pca.fit_transform(data)
```

```python
newData.shape
```

```python
plt.scatter(newData[:,0],newData[:,1],c=y)
plt.show()
```
