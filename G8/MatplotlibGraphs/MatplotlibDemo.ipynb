```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('classic')
```

```python
x = np.arange(11)
y1 = 2*x+3
y2 = 4*x+5
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,'r--',label='Potato')
plt.plot(x,y2,'go',label='Onion')
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.title('Vegetable Prices',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.arange(11)
y1 = x**2
y2 = x**2.5
y3 = x**2.2
```

```python
plt.plot(x,y1,label='x**2')
plt.plot(x,y2,label='x**2.5')
plt.plot(x,y3,label='x**2.2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Mathematical Functions',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```
