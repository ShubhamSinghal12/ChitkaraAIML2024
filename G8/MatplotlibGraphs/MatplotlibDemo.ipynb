```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('classic')
```

```python
x = np.arange(11)
y1 = 2*x+3
y2 = 4*x+5
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,'r--',label='Potato')
plt.plot(x,y2,'g:o',label='Onion')
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.title('Vegetable Prices',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.arange(11)
y1 = x**2
y2 = x**2.5
y3 = x**2.2
```

```python
plt.plot(x,y1,label='x**2')
plt.plot(x,y2,label='x**2.5')
plt.plot(x,y3,label='x**2.2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Mathematical Functions',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
plt.plot(x,y1,label='Potato',color = 'r',linestyle="--",linewidth=2
         ,marker='*',markerfacecolor = 'b',markersize=10)
plt.plot(x,y2,'g:o',label='Onion')
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.title('Vegetable Prices',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
# plt.plot(x,y1,'r--',x,y2,'g:^')
lines = plt.plot(x,y1,x,y2)
plt.setp(lines[0],label='Potato',color = 'r',linestyle="--",linewidth=2
         ,marker='*',markerfacecolor = 'b',markersize=10)
plt.setp(lines[1],'color','g',label='Onion')
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.title('Vegetable Prices',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.array([1,2,3])*2
y1 = np.array([10,20,30])
y2 = np.array([13,22,25])

plt.bar(x,y1,tick_label = ['Silver','Gold','Platinum'],width=0.5,label="Previous Year")
plt.bar(x+0.5,y2,width=0.5,label="Current Year")
plt.ylabel('Prices')
plt.title("Metal Prices")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3])*2
y1 = np.array([10,20,30])
y2 = np.array([13,22,25])

plt.barh(x,y1,tick_label = ['Silver','Gold','Platinum'],label="Previous Year")
plt.barh(x+0.5,y2,label="Current Year")
# plt.ylabel('Prices')
plt.title("Metal Prices")
plt.legend()
plt.show()
```

```python
x = np.linspace(0,5,1000)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

plt.figure(figsize=(10,4))

plt.subplot(211)
plt.plot(x,y1)
plt.title("Sin Graph")

plt.subplot(212)
plt.plot(x,y2)
plt.title("Cos Graph")

plt.suptitle('Trignometric Functions')
plt.subplots_adjust(hspace=1)
plt.show()
```

```python
x = np.linspace(0,5,1000)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

plt.figure(figsize=(10,4))

ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.plot(x,y1)
ax1.set_title("Sin Graph")

ax2.plot(x,y2)
ax2.set_title("Cos Graph")

plt.suptitle('Trignometric Functions')
plt.subplots_adjust(hspace=1)
plt.show()
```

```python
x = np.linspace(0,5,1000)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

# plt.figure(figsize=(10,4))
fig,ax = plt.subplots(2,2,figsize=(10,4))

ax[0,0].plot(x,y1)

ax[0,0].set_title("Sin Graph")

ax[0,1].plot(x,y2)
ax[0,1].set_title("Cos Graph")

fig.suptitle('Trignometric Functions')
fig.subplots_adjust(hspace=0.5,wspace=0.5)
plt.show()
```
