```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.groupby('day').mean(True)
```

```python
sns.barplot(data=tips, y = 'day',x='total_bill',estimator='sum',
            hue='sex',orient='h')
```

```python
sns.histplot(data=tips,x='total_bill',kde=True)
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(data=fmri,x='timepoint',y='signal',style='event',marker=True,dashes=False,hue='region')
```
