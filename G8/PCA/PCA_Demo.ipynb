```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
```

```python
X = np.random.randint(1,100,5000).reshape((1000,5))
k = 2
X.shape
```

```python
data = load_breast_cancer()
X = data.data
scale = StandardScaler()
X = scale.fit_transform(X)
y = data.target
k = 3
```

```python
covM = np.cov(X,rowvar=False)
covM.shape
```

```python
eigenValues, eigenVectors = np.linalg.eig(covM)
```

```python
eigenValues,eigenValues.shape
```

```python
# eigenVectors,
eigenVectors.shape
```

```python
sorted_index = np.argsort(eigenValues)[::-1]
sorted_index
```

```python
sorted_Values = eigenValues[sorted_index]
# sorted_Values
np.cumsum(sorted_Values)/np.sum(sorted_Values)
```

```python
sorted_Vectors = eigenVectors[sorted_index]
# sorted_Vectors
```

```python
bestK = sorted_Vectors[:k,:]
bestK.shape
```

```python
newData = (bestK.dot(X.T)).T
```

```python
newData.shape
```

```python
plt.scatter(newData[:,0],newData[:,1],c=y)
plt.show()
```
